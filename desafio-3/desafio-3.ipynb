## Escape HTML
#### Es un filtro de seguridad, que evita que un usuario pueda romper la página o robar información.

```python
%%file escape.py
from flask import Flask
from markupsafe import escape

app = Flask(__name__)

@app.route('/<name>')
def hello(name):
    return f'Hello {escape(name)}'
```

```python
!set FLASK_APP=escape.py
```

```python
!python -m flask --app escape run
```

##### Servidor: http://127.0.0.1:5000/Grupo5

## Enrutamiento

#### Sirve para conectar una dirección URL específica con una función de Python.

```python
%%file enrutamiento.py
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'Index Page'

@app.route('/hello')
def hello():
    return 'Hello, World'
```

```python
!set FLASK_APP=enrutamiento.py
```

```python
!python -m flask --app enrutamiento run
```

##### Servidor: http://127.0.0.1:5000/ o http://127.0.0.1:5000/hello
